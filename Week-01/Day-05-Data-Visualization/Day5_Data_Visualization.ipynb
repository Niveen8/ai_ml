```python
years = range(2000,2008)
apples =[0.895,0.91,0.919,0.926,0.923,0.936,0.901,0.981]
orangees =[0.962,0.941,0.930,0.918,0.908,0.907,0.904,0.909]
```

```python
print(type(years))
print(len(years))
print(list(years))

print(type(apples))
print(len(apples))
print(apples)
```

```python
plt.plot(years, apples)
plt.plot(years, orangees)
plt.xlabel('years')
plt.ylabel('yield (tons per hactare)');
```

```python
plt.plot(years, apples)
plt.plot(years, orangees)
plt.xlabel('years')
plt.ylabel('yield (tons per hactare)')
plt.title("crop yields in kanto")
plt.legend(['Apples','orangees']);
```

```python
plt.plot(years, apples ,marker='o')
plt.plot(years, orangees,marker='x')
plt.xlabel('years')
plt.ylabel('yield (tons per hactare)')
plt.title("crop yields in kanto")
plt.legend(['Apples','orangees']);
```

```python
plt.plot(years, apples ,marker='s')
plt.plot(years, orangees,marker='v')
plt.xlabel('years')
plt.ylabel('yield (tons per hactare)')
plt.title("crop yields in kanto")
plt.legend(['Apples','orangees']);
```

```python
plt.plot(years, apples ,marker='o',c='b')
plt.plot(years, orangees,marker='x',c='r')
plt.xlabel('years')
plt.ylabel('yield (tons per hactare)')
plt.title("crop yields in kanto")
plt.legend(['Apples','orangees']);
```

```python
plt.plot(years, apples ,marker='o',c='b',ls='-')
plt.plot(years, orangees,marker='x',c='r',ls='--')
plt.xlabel('years')
plt.ylabel('yield (tons per hactare)')
plt.title("crop yields in kanto")
plt.legend(['Apples','orangees']);
```

```python
plt.plot(years, apples ,marker='o',c='b',ls='-',ms=8)
plt.plot(years, orangees,marker='x',c='r',ls='--',ms=10)
plt.xlabel('years')
plt.ylabel('yield (tons per hactare)')
plt.title("crop yields in kanto")
plt.legend(['Apples','orangees']);
```

```python
plt.plot(years, apples ,marker='o',c='b',ls='-',ms=8,mew=2)
plt.plot(years, orangees,marker='x',c='r',ls='--',ms=10,alpha=.5)
plt.xlabel('years')
plt.ylabel('yield (tons per hactare)')
plt.title("crop yields in kanto")
plt.legend(['Apples','orangees']);
```

```python
plt.plot(years, apples ,marker='o',c='b',ls='-',ms=8,mew=2,mec='navy')
plt.plot(years, orangees,marker='x',c='r',ls='--',ms=10,alpha=.5)
plt.xlabel('years')
plt.ylabel('yield (tons per hactare)')
plt.title("crop yields in kanto")
plt.legend(['Apples','orangees']);
```

```python
plt.plot(years, apples ,'s-b')
plt.plot(years, orangees,'o--r')
plt.xlabel('years')
plt.ylabel('yield (tons per hactare)')
plt.title("crop yields in kanto")
plt.legend(['Apples','orangees']);
```
