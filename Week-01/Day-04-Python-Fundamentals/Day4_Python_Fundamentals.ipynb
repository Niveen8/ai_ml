4.4  Cleaning Data

```python
df.isnull().sum()
```

```python
df["sepal_length"].fillna(df["sepal_length"].mean(), inplace=True)
```

```python
df["sepal_length"].fillna(df["sepal_length"].mean(), inplace=False)
```

```python
df["petal_width"].fillna(df["petal_width"].mean(), inplace=True)
```

```python
df["petal_width"].fillna(df["petal_width"].mean(), inplace=False)
```

```python

df.dropna(inplace=True)
```

```python
df.dropna(inplace=False)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.drop_duplicates(inplace=False)
```

4.5  Grouping and Aggregation

```python
df.groupby("species")["petal_length"].mean()
```

```python
df.groupby("petal_width")["petal_length"].mean()
```

```python
df.groupby("petal_width")["petal_length"].max()
```

```python
df.groupby("petal_width")["petal_length"].min()
```

```python
df.groupby("species").agg({ "sepal_length": "mean","petal_length": "max"})
```

```python
df.groupby("species").agg({ "petal_width": "mean","petal_length": "max"})
```

```python
df.groupby("species").agg({ "petal_width": "min","petal_length": "max"})
```

```python
df.agg({ "petal_width": "min","petal_length": "max"})
```
