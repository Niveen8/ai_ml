2.5 Object-Oriented Programming (Basics)


```python
class DataPoint:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def distance_from_origin(self):
        return (self.x**2 + self.y**2) ** 0.5
p = DataPoint(3, 4)
print(p.distance_from_origin())
```

```python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade
    def show_grade(self):
        return self.grade
s = Student("Niveen", 90)
print(s.show_grade())
```

## Step 1: Write a function that takes a list of numbers and returns the mean, min, and max as a dictionary
==> max
==>min
==>mean

```python
def calculate_statistics(numbers):
    return {
        "mean": sum(numbers) / len(numbers),
        "min": min(numbers),
        "max": max(numbers)
    }
numbers = [100, 120, 130, 140, 150]
print(calculate_statistics(numbers))
```
